# 冒泡排序

```python
def bubble(nums):
    for i in range(1,len(nums)):
        for j in range(0,len(nums) - i):
            if nums[j] > nums[j+1]:
                nums[j], nums[j+1] = nums[j+1], nums[j]
    return nums

nums = [2,3,1,8,5]
bubble(nums)
```

# 快速排序

```python
def quick_sort(nums):
    if len(nums) <= 1:
        return nums
    p = nums[-1]
    left = [x for x in nums[:-1] if x <= p]
    right = [x for x in nums[:-1] if x > p]

    return quick_sort(left) + [p] + quick_sort(right)

nums = [2,3,1,8,5]
quick_sort(nums)
```

# 选择排序

```python
def select_sort(nums):
    for i in range(len(nums)):
        min_idx = i
        for j in range(i+1,len(nums)):
            if nums[j] < nums[min_idx]:
                min_idx = j
        nums[i],nums[min_idx] = nums[min_idx], nums[i]
    return nums
nums = [2,3,1,8,5]
select_sort(nums)
```

# 插入排序

```python
def insert_sort(nums):
    for i in range(1,len(nums)):
        key = nums[i]
        j = i - 1
        while j>=0 and key < nums[j]:
            nums[j+1] = nums[j]
            j -= 1
        nums[j+1] = key
    return nums
nums = [2,3,1,8,5]
insert_sort(nums)
```

# 求区间的质数

```python
def is_prime(x):
    if x < 2:
        return False
    i = 2
    while i * i <= x:
        if x % i == 0:
            return False
        i += 1
    return True
ans = []
for i in range(1,20):
    if is_prime(i):
        ans.append(i)
print(ans)
```

# 求开平方

牛顿迭代法（收敛更快，工业常用）数学原理
求$\sqrt{num}$，等价于解方程 $f(x) = x^2 - num = 0$

牛顿迭代公式：

$$x_{n+1} = \frac{1}{2}\left(x_n + \frac{num}{x_n}\right)$$
不断迭代，直到两次结果差值小于精度，收敛得到平方根。

```python
def sqrt_(n):
    if n < 0:
        return None
    x = n
    for _ in range(100):  # 固定迭代次数，保证精度
        x = (x + n / x) / 2
    return x

# 调用
print(sqrt_(2))
```

二分查找法
原理

对于非负数 num：

若 0 < num < 1：平方根在 [num, 1] 之间

若 num ≥ 1：平方根在 [0, num] 之间

不断二分区间，缩小范围，直到区间长度小于设定精度，输出近似值。

```python
def my_sqrt(n):
    if n < 0:
        return "负数无实数根"
    eps = 1e-6  # 精度，考试常用 1e-6 / 1e-8
    l, r = (n, 1.0) if n < 1 else (0.0, float(n))
    while r - l > eps:
        mid = (l + r) / 2
        if mid * mid < n:
            l = mid
        else:
            r = mid
    return (l + r) / 2

# 测试调用
print(my_sqrt(2))
```

# 二叉树的中序遍历（使用栈来实现）

```python
def inorderTraversal(root):
    stack = []
    res = []
    cur = root
    while cur or stack:
        # 先一路向左走，把路上的节点都压栈
        while cur:
            stack.append(cur)
            cur = cur.left
        # 左边走到底了，弹出栈顶节点
        cur = stack.pop()
        res.append(cur.val)
        # 再去右子树
        cur = cur.right
    return res
```

# 构建链表，打印链表

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def bulidListNode(nums):
    if not nums:
        return None
    dummy = ListNode(0)
    cur = dummy
    for i in range(0,len(nums)):
        cur.next = ListNode(nums[i])
        cur = cur.next
    return dummy.next

def printListNode(root):
    if not root:
        return []
    ans = []
    while root:
        ans.append(root.val)
        root = root.next
    print(ans)

nums = [1,2,3]

printListNode(bulidListNode(nums))
```
